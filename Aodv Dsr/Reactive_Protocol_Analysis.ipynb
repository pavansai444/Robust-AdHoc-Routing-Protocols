```python
import numpy as np
import pandas as pd
from scipy.spatial import distance
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

def read_data(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    data_DSR = {'n': [], 'Throughput': [], 'End-to-End Delay': [], 'Packet Drop %': [], 'Routing Load': []}
    current_n = None
    divide_n = [2,5,7,9,38,48,76]
    i=0;
    for line in lines:
        if line.startswith("#------------------n="):
            current_n = int(line.split('=')[1].split('-')[0].strip())
        elif line.startswith("Average Throughput is ="):
            throughput = float(line.split('=')[1].split()[0])
            throughput = throughput/(divide_n[i])
            i = i+1
            data_DSR['n'].append(current_n)
            data_DSR['Throughput'].append(throughput)
        elif line.startswith("Average end to end delay(ms)="):
            end_to_end_delay = float(line.split('=')[1].strip())
            data_DSR['End-to-End Delay'].append(end_to_end_delay)
        elif line.startswith("packet drop % ="):
            packet_drop = float(line.split('=')[1].strip())
            data_DSR['Packet Drop %'].append(packet_drop)
        elif line.startswith("Normalised Routing Load ="):
            routing_load = float(line.split('=')[1].strip())
            data_DSR['Routing Load'].append(routing_load)
    
    return data_DSR

def plot_graph(x_values, y_values, x_label, y_label, title):
    plt.plot(x_values, y_values, marker='o')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()

file_path = 'throughput_DSR.txt'
data_DSR = read_data(file_path)

# Plotting Throughput
plot_graph(data_DSR['n'], data_DSR['Throughput'], 'n', 'Average Throughput (kbps)', 'DSR with TCP Average Throughput vs n')

# Plotting End-to-End Delay
plot_graph(data_DSR['n'], data_DSR['End-to-End Delay'], 'n', 'End-to-End Delay (ms)', 'DSR with TCP End-to-End Delay vs n')

# Plotting Packet Drop %
plot_graph(data_DSR['n'], data_DSR['Packet Drop %'], 'n', 'Packet Drop %', 'DSR with TCP Packet Drop % vs n')

# Plotting Routing Load
plot_graph(data_DSR['n'], data_DSR['Routing Load'], 'n', 'Routing Load', 'DSR with TCP Routing Load vs n')
```
